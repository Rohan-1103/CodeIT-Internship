### 1. Print dynamic star pattern for a given n
> WebKul – Python Developer

```python
def print_star_pattern(n):
    if not isinstance(n, int) or n <= 0:
        print("Please provide a positive integer for the height.")
        return

    for i in range(n):
        print(' ' * (n - i - 1) + '*' * (2 * i + 1))

try:
    num = int(input("Enter the height of the star pattern: "))
    print_star_pattern(num)
except ValueError:
    print("Invalid input. Please enter an integer.")
```

### 2. Read a nested list & flatten it
> General interview task style

```python
def flatten_list(nested_list):
    """Flattens a nested list using recursion."""
    flat_list = []
    for item in nested_list:
        if isinstance(item, list):
            flat_list.extend(flatten_list(item))
        else:
            flat_list.append(item)
    return flat_list

nested = [1, [2, 3], [4, [5, 6]], 7]
print(f"Nested list: {nested}")
print(f"Flattened list: {flatten_list(nested)}")
```

### 4. Sort dictionary by values
> General interview task style

```python
def sort_dict_by_value(d):
    return {k: v for k, v in sorted(d.items(), key=lambda item: item[1])}

my_dict = {'apple': 3, 'banana': 1, 'cherry': 2}
print(f"Original dictionary: {my_dict}")
sorted_dict = sort_dict_by_value(my_dict)
print(f"Sorted dictionary: {sorted_dict}")
```

### 5. Program to count vowels, digits, etc., in a string
> GajShield Infotech – Python Developer

```python
def count_chars(s):
    vowels = 'aeiouAEIOU'
    digits = '0123456789'    
    v_count, d_count, c_count, s_count = 0, 0, 0, 0
    for char in s:
        if char in vowels:
            v_count += 1
        elif char in digits:
            d_count += 1
        elif char.isalpha():
            c_count += 1
        else:
            s_count += 1
    return {'vowels': v_count, 'digits': d_count, 'consonants': c_count, 'special_chars': s_count}

text = "Hello World! 123"
counts = count_chars(text)
print(f"In the string '{text}':")
print(f"- Vowels: {counts['vowels']}")
print(f"- Digits: {counts['digits']}")
print(f"- Consonants: {counts['consonants']}")
print(f"- Special Characters: {counts['special_chars']}")
```

### 6. Implement OOP (inheritance, polymorphism)
> Morgan Stanley – Software Developer

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Subclass must implement abstract method")

class Dog(Animal): # Inheritance
    def speak(self):
        return f"{self.name} says Woof!"

class Cat(Animal): # Inheritance
    def speak(self):
        return f"{self.name} says Meow!"

# Polymorphism
animals = [Dog("Buddy"), Cat("Whiskers")]

for animal in animals:
    print(animal.speak())
```

### 7. Write a generator to yield even numbers up to n
> Typical interview coding (Python OOP / generator)

```python
def even_generator(n):
    for i in range(n + 1):
        if i % 2 == 0:
            yield i

print("Even numbers up to 10:")
for num in even_generator(10):
    print(num)
```

### 8. Function to remove duplicates from a list
> Standard Python task

```python
def remove_duplicates(lst):
    return list(dict.fromkeys(lst))

# Example usage:
my_list = [1, 2, 3, 2, 1, 4, 5, 4]
print(f"Original list: {my_list}")
unique_list = remove_duplicates(my_list)
print(f"List with duplicates removed: {unique_list}")
```
